```python
import os
import json
from datetime import datetime

DATA_FILE = "students.json"
REPORT_FILE = "report.txt"


class Student:
    """Represents a single student record."""

    def __init__(self, student_id, name, age, course):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.course = course

    def to_dict(self):
        return {
            "student_id": self.student_id,
            "name": self.name,
            "age": self.age,
            "course": self.course,
        }

    def __str__(self):
        return f"ID: {self.student_id} | Name: {self.name} | Age: {self.age} | Course: {self.course}"


class StudentManager:
    """Handles all CRUD operations and persistence for student records."""

    def __init__(self):
        self.students = []
        self.added = 0
        self.updated = 0
        self.deleted = 0
        self.load_data()

    def load_data(self):
        if not os.path.exists(DATA_FILE):
            with open(DATA_FILE, "w") as f:
                json.dump([], f)

        try:
            with open(DATA_FILE, "r") as f:
                data = json.load(f)
                self.students = [Student(**item) for item in data]
        except (json.JSONDecodeError, TypeError):
            print("Warning: data file was invalid or empty. Starting with an empty list.")
            self.students = []

    def save_data(self):
        with open(DATA_FILE, "w") as f:
            json.dump([s.to_dict() for s in self.students], f, indent=4)

    def find_student(self, student_id):
        for student in self.students:
            if student.student_id == student_id:
                return student
        return None

    def add_student(self, student_id, name, age, course):
        if self.find_student(student_id):
            print("A student with this ID already exists.")
            return
        self.students.append(Student(student_id, name, age, course))
        self.save_data()
        self.added += 1
        print("Student added.")

    def view_students(self):
        if not self.students:
            print("No student records found.")
            return
        print(f"\n{'ID':<10}{'Name':<20}{'Age':<6}{'Course':<15}")
        print("-" * 51)
        for student in self.students:
            print(f"{student.student_id:<10}{student.name:<20}{str(student.age):<6}{student.course:<15}")

    def search_student(self, student_id):
        student = self.find_student(student_id)
        if student:
            print(student)
        else:
            print("Student not found.")
        return student

    def update_student(self, student_id, name=None, age=None, course=None):
        student = self.find_student(student_id)
        if not student:
            print("Student not found.")
            return
        if name:
            student.name = name
        if age:
            student.age = age
        if course:
            student.course = course
        self.save_data()
        self.updated += 1
        print("Student updated.")

    def delete_student(self, student_id):
        student = self.find_student(student_id)
        if not student:
            print("Student not found.")
            return
        self.students.remove(student)
        self.save_data()
        self.deleted += 1
        print("Student deleted.")

    def save_report(self):
        with open(REPORT_FILE, "w") as f:
            f.write("=" * 50 + "\n")
            f.write("STUDENT RECORD MANAGER - SUMMARY REPORT\n")
            f.write("=" * 50 + "\n")
            f.write(f"Timestamp: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
            f.write(f"Total records currently stored: {len(self.students)}\n")
            f.write(f"Added this session   : {self.added}\n")
            f.write(f"Updated this session : {self.updated}\n")
            f.write(f"Deleted this session : {self.deleted}\n")
            f.write("=" * 50 + "\n")
        print(f"Report saved to {REPORT_FILE}")
```
