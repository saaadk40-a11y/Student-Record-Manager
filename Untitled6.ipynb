```python
import os
import json
from datetime import datetime

DATA_FILE = "students.json"
REPORT_FILE = "report.txt"


class Student:
    """Represents a single student record."""

    def __init__(self, student_id, name, age, course):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.course = course

    def to_dict(self):
        return {
            "student_id": self.student_id,
            "name": self.name,
            "age": self.age,
            "course": self.course,
        }

    def __str__(self):
        return f"ID: {self.student_id} | Name: {self.name} | Age: {self.age} | Course: {self.course}"


class StudentManager:
    """Handles all CRUD operations and persistence for student records."""

    def __init__(self):
        self.students = []
        self.added = 0
        self.updated = 0
        self.deleted = 0
        self.load_data()

    def load_data(self):
        if not os.path.exists(DATA_FILE):
            with open(DATA_FILE, "w") as f:
                json.dump([], f)

        try:
            with open(DATA_FILE, "r") as f:
                data = json.load(f)
                self.students = [Student(**item) for item in data]
        except (json.JSONDecodeError, TypeError):
            print("Warning: data file was invalid or empty. Starting with an empty list.")
            self.students = []

    def save_data(self):
        with open(DATA_FILE, "w") as f:
            json.dump([s.to_dict() for s in self.students], f, indent=4)

    def find_student(self, student_id):
        for student in self.students:
            if student.student_id == student_id:
                return student
        return None

    def add_student(self, student_id, name, age, course):
        if self.find_student(student_id):
            print("A student with this ID already exists.")
            return
        self.students.append(Student(student_id, name, age, course))
        self.save_data()
        self.added += 1
        print("Student added.")

    def view_students(self):
        if not self.students:
            print("No student records found.")
            return
        print(f"\n{'ID':<10}{'Name':<20}{'Age':<6}{'Course':<15}")
        print("-" * 51)
        for student in self.students:
            print(f"{student.student_id:<10}{student.name:<20}{str(student.age):<6}{student.course:<15}")

    def search_student(self, student_id):
        student = self.find_student(student_id)
        if student:
            print(student)
        else:
            print("Student not found.")
        return student

    def update_student(self, student_id, name=None, age=None, course=None):
        student = self.find_student(student_id)
        if not student:
            print("Student not found.")
            return
        if name:
            student.name = name
        if age:
            student.age = age
        if course:
            student.course = course
        self.save_data()
        self.updated += 1
        print("Student updated.")

    def delete_student(self, student_id):
        student = self.find_student(student_id)
        if not student:
            print("Student not found.")
            return
        self.students.remove(student)
        self.save_data()
        self.deleted += 1
        print("Student deleted.")

    def save_report(self):
        with open(REPORT_FILE, "w") as f:
            f.write("=" * 50 + "\n")
            f.write("STUDENT RECORD MANAGER - SUMMARY REPORT\n")
            f.write("=" * 50 + "\n")
            f.write(f"Timestamp: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
            f.write(f"Total records currently stored: {len(self.students)}\n")
            f.write(f"Added this session   : {self.added}\n")
            f.write(f"Updated this session : {self.updated}\n")
            f.write(f"Deleted this session : {self.deleted}\n")
            f.write("=" * 50 + "\n")
        print(f"Report saved to {REPORT_FILE}")


def get_non_empty(prompt):
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("This field cannot be empty.")


def get_valid_age(prompt, allow_blank=False):
    while True:
        value = input(prompt).strip()
        if allow_blank and value == "":
            return None
        if value.isdigit() and 0 < int(value) < 120:
            return int(value)
        print("Please enter a valid age (a number between 1 and 119).")


def run_student_manager():
    manager = StudentManager()

    while True:
        print("\n1. Add student\n2. View all\n3. Search\n4. Update\n5. Delete\n6. Exit")
        choice = input("Enter choice: ").strip()

        if choice == "1":
            student_id = get_non_empty("ID: ")
            name = get_non_empty("Name: ")
            age = get_valid_age("Age: ")
            course = get_non_empty("Course: ")
            manager.add_student(student_id, name, age, course)

        elif choice == "2":
            manager.view_students()

        elif choice == "3":
            student_id = get_non_empty("Enter ID to search: ")
            manager.search_student(student_id)

        elif choice == "4":
            student_id = get_non_empty("Enter ID to update: ")
            if not manager.find_student(student_id):
                print("Student not found.")
                continue
            name = input("New name (blank to skip): ").strip()
            age = get_valid_age("New age (blank to skip): ", allow_blank=True)
            course = input("New course (blank to skip): ").strip()
            manager.update_student(student_id, name, age, course)

        elif choice == "5":
            student_id = get_non_empty("Enter ID to delete: ")
            manager.delete_student(student_id)

        elif choice == "6":
            manager.save_report()
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 6.")


# Run the program (this line executes automatically when the cell runs)
run_student_manager()
```
